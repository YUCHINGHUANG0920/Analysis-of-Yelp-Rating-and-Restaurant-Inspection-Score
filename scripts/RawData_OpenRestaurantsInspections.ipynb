Import the library

```python
import pandas as pd
```

Read the whole dataset - "Open Restaurants Inspections"

```python
offset = 0
limit = 1000
```

```python
raw_data_RestInspect = pd.DataFrame()

while True:
    url = f'https://data.cityofnewyork.us/resource/4dx7-axux.json?$limit={limit}&$offset={offset}'
    df_chunk = pd.read_json(url)
    raw_data_RestInspect = pd.concat([raw_data_RestInspect, df_chunk], ignore_index=True)

    if len(df_chunk) < limit:
        break

    offset += limit
```

Clean the dataset

```python
cleaned_RestInspect = raw_data_RestInspect.copy()
```
